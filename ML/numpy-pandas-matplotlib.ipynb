```python
import time
import numpy as np

import functions_vectorised
import functions
```

```python
#  

sum_non_neg_diag_data = []
for n in [10, 100, 1000, 10000]:
    X = np.random.randint(0, 3, (n, n))
    sum_non_neg_diag_data += [X]

are_multisets_equal_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(0, 2, n)
    y = np.random.randint(0, 2, n)
    are_multisets_equal_data += [(x, y)]

max_prod_mod_3_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(1, 100, n)
    max_prod_mod_3_data += [x]

convert_image_data = []
for n in [10, 100, 1000]:
    image = np.random.random((n, n, 10))
    weights = np.random.random(10)
    convert_image_data += [(image, weights)]

rle_scalar_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(1, 3, (n, 2))
    y = np.random.randint(1, 3, (n, 2))
    shapes = np.random.randint(1, 3, n)
    x[:, 1] = shapes
    y[:, 1] = shapes

    rle_scalar_data += [(x, y)]

cosine_distance_data = []
for n in [10, 100]:
    X = np.random.random((n, n))
    Y = np.random.random((n, n))
    cosine_distance_data += [(X, Y)]
```

```python
#     

def count_time(func, *data):
    tick = time.perf_counter()
    func(*data)
    return time.perf_counter() - tick
```

```python
#    

# sum_non_neg_diag

sum_non_neg_diag_time = []
sum_non_neg_diag_vectorised_time = []

for item in sum_non_neg_diag_data:
   sum_non_neg_diag_time += [count_time(functions.sum_non_neg_diag, item.tolist())]
   sum_non_neg_diag_vectorised_time += [count_time(functions_vectorised.sum_non_neg_diag, item)]

# are_multisets_equal
are_multisets_equal_time = []
are_multisets_equal_vectorised_time = []

for item in sum_non_neg_diag_data:
   are_multisets_equal_time += [count_time(functions.are_multisets_equal, item[0].tolist(), item[1].tolist())]
   are_multisets_equal_vectorised_time += [count_time(functions_vectorised.are_multisets_equal, item[0], item[1])]

# max_prod_mod_3
max_prod_mod_3_time = []
max_prod_mod_3_vectorised_time = []

for item in max_prod_mod_3_data:
   max_prod_mod_3_time += [count_time(functions.max_prod_mod_3, item.tolist())]
   max_prod_mod_3_vectorised_time += [count_time(functions_vectorised.max_prod_mod_3, item)]

# convert_image
convert_image_time = []
convert_image_vectorised_time = []

for item in convert_image_data:
   convert_image_time += [count_time(functions.convert_image, item[0].tolist(), item[1].tolist())]
   convert_image_vectorised_time += [count_time(functions_vectorised.convert_image, item[0], item[1])]

# rle_scalar
rle_scalar_time = []
rle_scalar_vectorised_time = []

for item in rle_scalar_data:
   rle_scalar_time += [count_time(functions.rle_scalar, item[0].tolist(), item[1].tolist())]
   rle_scalar_vectorised_time += [count_time(functions_vectorised.rle_scalar, item[0], item[1])]

# cosine_distance
cosine_distance_time = []
cosine_distance_vectorised_time = []

for item in cosine_distance_data:
   cosine_distance_time += [count_time(functions.cosine_distance, item[0].tolist(), item[1].tolist())]
   cosine_distance_vectorised_time += [count_time(functions_vectorised.cosine_distance, item[0], item[1])]
```

```python
fig, axs = plt.subplots(figsize=(16, 9),nrows=2, ncols=3)
fig.suptitle('     ')

func = [['sum_non_neg_diag', 'are_multisets_equal', 'max_prod_mod_3'], ['convert_image', 'rle_scalar', 'cosine_distance']]

for i in range(2):
  for j in range(3):
    data_simple   = eval(func[i][j]+'_time')
    axs[i, j].plot([10 ** (i + 1) for i in range(len(data_simple))], data_simple, label='non vectorised')
    data_optimize = eval(func[i][j]+'_vectorised_time')
    axs[i, j].plot([10 ** (i + 1) for i in range(len(data_optimize))], data_optimize, label='vectorised')

    axs[i, j].set_xscale('log')
    axs[i, j].set_xlabel(' ')
    axs[i, j].set_yscale('log')
    axs[i, j].set_ylabel(' ')
    axs[i, j].set_title(func[i][j])
    axs[i, j].legend(loc="upper left")
plt.show()
```
